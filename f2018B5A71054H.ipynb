Question:

 Estimate the integral of 4*(1-x^2)^0.5 in the interval 0 ≤ x ≤ 1 using the Monte-Carlo (hit or miss) method. .Program should calculate the theoretical value of the error in the evaluation of the integral, i.e, it should calculate the standard deviation of the function and the standard deviation of the integral values determined. 

That is evaluate the integral by taking 1000 points and repeat this 1000 times. Calculate the Std. deviation of these 1000 results and also the std. deviation of the function evaluations.

 

Calculate the actual error in your result, i.e. take the difference between your numerical result with the actual value of the integral. The actual value of the integral can be obtained analytically  </p>

In this assignment i shall refer area as the value of the definite intergral
****************************************************************************

```
%matplotlib inline 
from scipy import random
import  numpy as np
import matplotlib.pyplot as plt
```

```
a=0 #lower limit of integral
b=1; #upper limit of intergral
N=1000 #number of points taken 
M=1000 # number of times the monte carlo simulation must be run
```

```
#returns the function value 
def func(x):
  return 4*((1-x**2)**0.5)

```

```
areas=[] #list consists of all N areas/integral calculated 
f=[]#holds all the function values
#Monte Carlo simulation 
for i in range(M):
  xrand= np.random.uniform(low=a, high=b, size=(1000,)) #array with N random numbers from a to b i.e (0,1)
  fval=0.0 
  for i in range(N):
    x=func(xrand[i])
    f.append(x)
    fval+=x

  answer=(b-a)*fval/float(N) #answer= value of the definite integral in each loop
  areas.append(answer)


```

```


plt.title("Distribution of areas calculated")
plt.hist(areas,bins=50,ec="black")
plt.xlabel("areas")
```

```
s=np.std(areas)
m=np.mean(areas)
functionStandardDeviation=s*(N**.5);#SD of area = Sd of function/root N
print("the standard deviation of  4*(1-x^2)^0.5 calculated is:",np.std(f))
print("the standard deviation of  4*(1-x^2)^0.5 calculated is:",functionStandardDeviation)
print("approx area calculated",m)
print("standard deviation of areas calc :",s)
print("analytically calculated area:3.141592653589793")
print("error in area calculated:",3.141592653589793-m)
print(f"% error in area calculated:{abs((3.141592653589793-m)/3.141592653589793*100)}%")
```
